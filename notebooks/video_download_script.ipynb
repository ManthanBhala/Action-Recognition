```python
import os
import matplotlib
import numpy as np
import pandas as pd
import requests
import openpyxl
```

```python
def download_video(video_links,sessions): 
    i=0
    for link in video_links: 
        
        file_name = link.split('/')[-1]
        file_name = sessions[i] + "-" + file_name  
        dir="Data/Original/"
        print("Downloading file:%s"%file_name) 
        
        exists = os.path.isfile(dir + file_name)
        if exists==1:
            print("Already exists")
            i+=1
            continue
             
        
        r = requests.get(link, stream = True) 
          
        with open(dir + file_name, 'wb') as f: 
            for chunk in r.iter_content(chunk_size = 1024*1024): 
                if chunk: 
                    f.write(chunk) 
        print("%s downloaded!\n"%file_name)
        i+=1
```

```python
wb = openpyxl.load_workbook('dai-asllvd-BU_glossing_with_variations_HS_information-extended-urls-RU.xlsx')
ws = wb['Sheet1']
url=[]
sessions=[]
class_count=0
class_required=100
current_class=""
i=1
while(class_count<=class_required):
    
    l = ws.cell(row=i, column=18).value
    if(l[:10]!="=HYPERLINK"):
        i+=1
        continue
    
    session = ws.cell(row=i, column=13).value
    scene = ws.cell(row=i, column=14).value
    
    link = "http://csr.bu.edu/ftp/asl/asllvd/asl-data2/quicktime/" + session +  "/scene" + str(scene) + "-camera1.mov"
    
    
    class_name = ws.cell(row=i, column=4).value
    if(class_name!=current_class):
        current_class = class_name
        class_count+=1
    sessions.append(session)
    url.append(link)
    i+=1
    
```

```python
download_video(url,sessions)  
```
